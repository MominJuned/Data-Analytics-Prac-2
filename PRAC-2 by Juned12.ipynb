```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(10)

data = {
    "Month": np.random.choice(["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct"],100),
    "Region": np.random.choice(["North","South","East","West"],100),
    "Sales": np.random.randint(15000,50000,100),
    "Advertising_Spend": np.random.randint(2000,10000,100),
}

df = pd.DataFrame(data)

df["Profit"] = df["Sales"] - np.random.randint(8000,20000,100)

df.head()
```

```python
plt.scatter(df["Advertising_Spend"], df["Sales"],color="k")
plt.xlabel("Advertising Spend")
plt.ylabel("Sales")
plt.title("Advertising vs Sales")
plt.show()
```

## Predictive Analytics

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[["Advertising_Spend"]]
y = df["Sales"]

model = LinearRegression()
model.fit(X,y)
```

```python
model.predict([[7000]])
```

```python
plt.scatter(df["Advertising_Spend"], df["Sales"],color="r")

plt.plot(df["Advertising_Spend"], model.predict(X))

plt.xlabel("Advertising Spend")
plt.ylabel("Sales")
plt.title("Predictive Model")
plt.show()
```

```python
df["Recommended_Action"] = np.where(df["Sales"] < 25000,
                                    "Increase Advertising",
                                    "Maintain Strategy")

df.head()
```
